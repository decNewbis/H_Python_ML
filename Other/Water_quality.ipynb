```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize']=[15,6]
```

```python
df = pd.read_csv('./Water Quality Testing.csv')
df.head()
```

```python
df = df.drop('Sample ID',axis=1)
df.head()
```

```python
r=3
c=2
it=1
for i in df.columns:
    plt.subplot(r,c,it)
    sns.kdeplot(df[i])
    plt.grid()
    it+=1
plt.tight_layout()
plt.show()
```
