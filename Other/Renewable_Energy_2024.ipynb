```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the dataset
# Display the first few rows of the dataset
df = pd.read_csv('./complete_renewable_energy_dataset.csv')
df.head()
```

```python
df.describe()
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
print(missing_percentage)
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=df, x='Country', y='Production (GWh)')
plt.title('Distribution of Energy Production by Country')
plt.xticks(rotation=90)
plt.ylabel('Production (GWh)')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(data=df, x='Year', y='Production (GWh)', hue='Energy Type')
plt.title('Energy Production Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Production (GWh)')
plt.legend(title='Energy Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(data=df, x='Country', y='Investments (USD)', errorbar=None)
plt.title('Investment in Renewable Energy by Country')
plt.xticks(rotation=90)
plt.ylabel('Investments (USD)')
plt.show()
```

```python
numeric_data = df.select_dtypes(include=[np.number])

corr_matrix = numeric_data.corr()

plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap of All Features')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(data=df, x='Country', y='Proportion of Energy from Renewables')
plt.title('Proportion of Energy from Renewables by Country')
plt.xticks(rotation=90)
plt.ylabel('Proportion of Energy from Renewables (%)')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(data=df, x='GDP', y='Energy Consumption', hue='Country', size='Population', sizes=(20, 200))
plt.title('GDP vs Energy Consumption')
plt.xlabel('GDP')
plt.ylabel('Energy Consumption')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='Energy Type')
plt.title('Distribution of Energy Types')
plt.xlabel('Energy Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(data=df, x='Innovation Index', y='Ease of Doing Business', hue='Country', style='Energy Type', size='GDP', sizes=(20, 200))
plt.title('Innovation Index vs Ease of Doing Business')
plt.xlabel('Innovation Index')
plt.ylabel('Ease of Doing Business')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
